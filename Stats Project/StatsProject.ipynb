```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the dataset

df = pd.read_csv(r'C:\eduo\stats\16288282\Stats Project\Storedata.csv', encoding='ISO-8859-1')




# Display the first few rows and summary statistics

print(df.head())
print(df.info())
print(df.describe())
```

```python

# Handle missing values
# Checking for missing values
print("Missing values in each column:")
print(df.isnull().sum())
     
# Convert data types if necessary

# Create any new columns needed for analysis
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Ship Year'] = df['Ship Date'].dt.year
df['Ship Month'] = df['Ship Date'].dt.month


print(df.head())
```

## Additional Statistical Analysis


```python
import scipy.stats as stats

# Perform the T-test for sales in different regions
region_sales = df.groupby('Region')['Sales'].mean()

region_east = df[df['Region'] == 'East']['Sales']
region_west = df[df['Region'] == 'West']['Sales']

t_stat, p_value = stats.ttest_ind(region_east, region_west)

print(f"T-test: T-statistic = {t_stat:.4f}, p-value = {p_value:.4f}")

contingency_table = pd.crosstab(df['Ship Mode'], df['Segment'])
chi2_stat, chi2_p_value, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square test: chi2 = {chi2_stat:.4f}, p-value = {chi2_p_value:.4f}", f"dof = {dof}")
```

```python
# Ensure the 'Order Date' column is in datetime format (if it's not already)
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# Create a new 'Year-Month' column
df['Year-Month'] = df['Order Date'].dt.to_period('M')

# Group by 'Year-Month' and sum the sales
sales_trends = df.groupby('Year-Month')['Sales'].sum()

# Plotting the sales trends
plt.figure(figsize=(12, 6))
sales_trends.plot(kind='line', color='#e80e0e', marker='o', linewidth=2)
plt.title('Sales Trends Over Time', fontsize=20)
plt.xlabel('Year-Month', fontsize=16)
plt.ylabel('Total Sales', fontsize=16)
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
# Plotting profit trends over time

profit_trends = df.groupby('Year-Month')['Profit'].sum()
plt.figure(figsize=(12, 6))
profit_trends.plot(kind='line', color='#e80e0e', marker='o', linewidth=2)
plt.title('Monthly Profit Trend', fontsize=20)
plt.xlabel('Year-Month', fontsize=16)
plt.ylabel('Profit', fontsize=16)
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['Season'] = df['Order Month'].apply(get_season)
seasonal_sales = df.groupby('Season')['Sales'].sum()
plt.figure(figsize=(8, 6))
seasonal_sales.plot(kind='bar', color=['#e80e0e', '#0ee820', '#e8c40e', '#0e24e8'])
plt.title('Seasonal Sales Trends', fontsize=20)
plt.xlabel('Season', fontsize=16)
plt.ylabel('Total Sales', fontsize=16)
plt.xticks(rotation=0)
plt.grid(True)
plt.show()





```

```python

# Analyzing seasonal sales trends
# Grouping sales data by month and year
# Plotting seasonal sales trends

seasonal_sales = df.groupby(['Order Year', 'Order Month'])['Sales'].sum().unstack()
plt.figure(figsize=(12, 8))
sns.heatmap(seasonal_sales, cmap='YlGnBu', annot=True, fmt='.0f', linewidths=.5)
plt.title('Seasonal Sales Trends by Year and Month', fontsize=16)
plt.xlabel('Order Month', fontsize=14)
plt.ylabel('Order Year', fontsize=14)
plt.show()
```

```python
# Comparing sales across different ship modes

sales_by_ship_mode = df.groupby('Ship Mode')['Sales'].sum()
plt.figure(figsize=(10, 6))
sales_by_ship_mode.plot(kind='bar', color='#3749e6', edgecolor='black')
plt.title('Total Sales Ship Modes', fontsize=20)
plt.xlabel('Ship Mode', fontsize=16)
plt.ylabel('Total Sales', fontsize=16)
plt.xticks(rotation=0)
plt.grid(True)
plt.show()

print("\nShip Modes:")
for ship_mode, total_sales in sales_by_ship_mode.items():
    print(f"{ship_mode}: {total_sales:,.2f}")





```
